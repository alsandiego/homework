```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time

# Import API key
from api_keys import api_key

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = 'https://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID='
#blank list to store retrieve data
city=[]
cloudiness=[]
country=[]
date=[]
humidity=[]
lat=[]
lng=[]
maxtemp=[]
windspeed=[]

#begin printing
print('Beginning Data Retrieval')
print('-----------------------------')

for i in range(len(cities)):
    #build the API url
    query_url = requests.get(url + api_key + "&q=" + cities[i]).json()
    
    #using exception handling to retrieve data
    try:
        print('Processing Record '+str(i)+' | '+str(query_url['name'])+' - '+
              str(query_url['id']))
        #print(url + api_key + "&q=" + cities[i])
        city.append(query_url['name'])
        cloudiness.append(query_url['clouds']['all'])
        country.append(query_url['sys']['country'])
        date.append(query_url['dt'])
        humidity.append(query_url['main']['humidity'])
        lat.append(query_url['coord']['lat'])
        lng.append(query_url['coord']['lon'])
        maxtemp.append(query_url['main']['temp_max'])
        windspeed.append(query_url['wind']['speed'])
        
    except KeyError:
        print("City not found. Skipping...")
print('-----------------------------')
print('Data Retrieval Complete')
print('-----------------------------')
```

### Convert Raw Data to DataFrame
* Export the city data into a .csv.
* Display the DataFrame

```python
#put all list into Dataframe
df_data = pd.DataFrame({'City':city, 'Cloudiness':cloudiness, 'Country':country,
                       'Date':date, 'Humidity':humidity, 'Lat':lat, 'Lng':lng,
                       'Max Temp':maxtemp, 'Wind Speed':windspeed})

df_data.head()
```

```python
#save dataframe to csv
df_data.to_csv(output_data_file)
```

```python
#check number of data
df_data.count()
```

### Plotting the Data
* Use proper labeling of the plots using plot titles (including date of analysis) and axes labels.
* Save the plotted figures as .pngs.

#### Latitude vs. Temperature Plot

```python
latvstemp = df_data[['Max Temp','Lat']]

graph = latvstemp.plot(kind='scatter', x='Lat', y= 'Max Temp',
                       figsize=(6,4),title="City Latitude vs. Max Temperature (10/1/2019)")
graph.xaxis.grid(color='gray')
graph.yaxis.grid(color='gray')
plt.xlabel('Latitude')
plt.ylabel('Max Temperature (F)')

plt.savefig("City Latitude vs Temperature.png")
plt.show()
```

#### Latitude vs. Humidity Plot

```python
latvshumidity = df_data[['Humidity','Lat']]
latvshumidity.plot.scatter('Lat','Humidity', s=40)

plt.title('City Latitude vs Humidity (10/01/2019)')
plt.xlabel('Latitude')
plt.ylabel('Humidity (%)')
plt.gca().xaxis.grid(True)
plt.gca().yaxis.grid(True)
plt.savefig("City Latitude vs Humidity.png")
plt.show()
```

#### Latitude vs. Cloudiness Plot

```python
latvscloud = df_data[['Cloudiness','Lat']]

graph = latvscloud.plot(kind='scatter', x='Lat', y= 'Cloudiness',
                       figsize=(6,4),title="City Latitude vs. Cloudiness (10/1/2019)")
graph.xaxis.grid(color='gray')
graph.yaxis.grid(color='gray')
plt.xlabel('Latitude')
plt.ylabel('Cloudiness (%)')

plt.savefig("City Latitude vs Cloudiness.png")
plt.show()
```

#### Latitude vs. Wind Speed Plot

```python
latvswind = df_data[['Wind Speed','Lat']]

graph = latvswind.plot(kind='scatter', x='Lat', y= 'Wind Speed',
                       figsize=(6,4),title="City Latitude vs. Wind Speed (10/1/2019)")
graph.xaxis.grid(color='gray')
graph.yaxis.grid(color='gray')
plt.xlabel('Latitude')
plt.ylabel('Wind Speed (mph)')

plt.savefig("City Latitude vs Wind Speed.png")
plt.show()
```
